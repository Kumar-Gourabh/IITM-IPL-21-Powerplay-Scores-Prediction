```python
# Importing the required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
pd.pandas.set_option('display.max_columns',None) # Display all the columns
```

```python
df=pd.read_csv('all_matches.csv')
df.head()
```

```python
df.shape
```

```python
# Let's take only First Six Overs data in each match
df = df[df['ball']<6.0]
df.head(50)
```

```python
df.shape
```

```python
df['venue'].unique()
```

```python
# Let's keep only current venues for IPL
current_venues=['Feroz Shah Kotla','Arun Jaitley Stadium','Sardar Patel Stadium, Motera',
                'M Chinnaswamy Stadium','Eden Gardens']
```

```python
df = df[(df['venue'].isin(current_venues))]
df['venue'].replace('Arun Jaitley Stadium', 'Feroz Shah Kotla', inplace=True)
```

```python
df.shape
```

```python
df['total'] = df.runs_off_bat+df.extras
```

```python
df.head()
```

```python
df.columns
```

```python
# Columns to remove
remove=['runs_off_bat', 'extras', 'wides', 'noballs', 'byes', 'legbyes',
       'penalty', 'wicket_type', 'player_dismissed', 'other_wicket_type',
       'other_player_dismissed']
df.drop(labels=remove, axis=1, inplace=True)
```

```python
df.head()
```

```python
# Also dropping ball as it's purpose was only to identify if the ball is within the first 6 overs
df.drop('ball', axis=1, inplace=True)
```

```python
df['batsmen']=df['striker']+', '+df['non_striker']
```

```python
df.drop(labels=['striker','non_striker'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['6_over_score'] = df.groupby(['match_id','innings'])['total'].transform(np.sum)
```

```python
df.drop('total', axis=1, inplace=True)
```

```python
df.head()
```

```python
df=df.assign(batsmen=df['batsmen'].str.split(',')).explode('batsmen')
```

```python
df.head()
```

```python
df.columns
```

```python
# Rearranging the columns
df = df[['match_id', 'season', 'start_date', 'venue', 'innings', 'batting_team',
       'bowling_team', 'batsmen', 'bowler', '6_over_score']]
```

```python
df=df.rename(columns={"bowler": "bowlers"})
```

```python
df.head()
```

```python
df['batsmen']=df['batsmen'].apply(lambda x: x.strip())
df['bowlers']=df['bowlers'].apply(lambda x: x.strip())
df['venue']=df['venue'].apply(lambda x: x.strip())
df=df.drop_duplicates()
df.head()
```

```python
df=df.drop(columns=['match_id','season'])
```

```python
df.head()
```

```python
df['batting_team'].unique()
```

```python
# Keeping only current teams
current_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                    'Mumbai Indians', 'Kings XI Punjab', 'Royal Challengers Bangalore',
                    'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
df = df[(df['batting_team'].isin(current_teams)) & (df['bowling_team'].isin(current_teams))]
```

```python
df.head()
```

```python
# Converting the column 'start_date' from string into datetime object
from datetime import datetime
df['start_date'] = df['start_date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
df_save=df.drop('start_date',axis=1)
```

```python
df_save=df_save.drop('6_over_score',axis=1)
```

```python
df_save.to_csv('File_for_Label_Encoding.csv',index=False) # Will use this to Label encode input data to to be predicted
```

```python
# Converting categorical features using LabelEncoder method

col=['venue','batting_team','bowling_team','batsmen','bowlers']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df[col] = df[col].apply(LabelEncoder().fit_transform)

df.head()
```

```python
# Splitting the data into train and test set based on year
X_train = df.drop(labels='6_over_score', axis=1)[df['start_date'].dt.year <= 2018]
X_test = df.drop(labels='6_over_score', axis=1)[df['start_date'].dt.year >= 2019]
```

```python
# Removing the 'date' column
X_train.drop(labels='start_date', axis=True, inplace=True)
X_test.drop(labels='start_date', axis=True, inplace=True)
```

```python
X_train.head()
```

```python
y_train = df[df['start_date'].dt.year <= 2018]['6_over_score'].values
y_test = df[df['start_date'].dt.year >= 2019]['6_over_score'].values
```

# Model Building

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='r2',cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
ridge_regressor=Ridge(alpha=40)
ridge_regressor.fit(X_train,y_train)
prediction=ridge_regressor.predict(X_test)
```

```python
import seaborn as sns
sns.distplot(y_test-prediction)
```

```python
from sklearn import metrics
import numpy as np
from sklearn.metrics import r2_score
r2 = r2_score(y_test, prediction)
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('R2', r2)
```

```python
prediction
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='r2',cv=5)

lasso_regressor.fit(X_train,y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
lasso_regressor=Lasso(alpha=35)
lasso_regressor.fit(X_train,y_train)
prediction=lasso_regressor.predict(X_test)
```

```python
import seaborn as sns
sns.distplot(y_test-prediction)
```

```python
r2 = r2_score(y_test, prediction)
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('R2', r2)
```

```python
prediction
```

# Random Forest Regression Model

```python
# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,50,100,150,200,250,500]}
rf = GridSearchCV(rfr, parameters, cv=5,scoring="r2")
rf.fit(X_train,y_train)
rf.best_params_
```

```python
rfr=RandomForestRegressor(n_estimators=200)
rfr.fit(X_train,y_train)
prediction=rfr.predict(X_test)
```

```python
sns.distplot(y_test-prediction)
```

```python
r2 = r2_score(y_test, prediction)
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('R2', r2)
```

```python
prediction
```

# Final Model - Lasso Regressor Gives Least R Squared Error, but seems to be overfitting. Random Forest gives a more Generalized model

```python
model=rfr
model.fit(X_train,y_train)
prediction=model.predict(X_test)
```

```python
# Creating a pickle file for the regressor
import joblib

joblib.dump(model,'6_over_score.pkl')
```
